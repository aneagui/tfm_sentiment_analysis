```python
import pandas as pd
```

```python
test_data = pd.read_json("test_data_LIMPIO.json")
```

```python
test_data.sample(15)
```

```python
grouped = test_data.groupby('Dataset')

# Paso 3: Crear un diccionario de DataFrames
tests = {name: group for name, group in grouped}
```

```python
amazon_test = tests["Amazon Review Polarity Dataset"]
yelp_test = tests["Yelp Polarity Reviews"]
financial_test = tests["Financial Phrasebank"]
stanford_test = tests["Stanford"]
sentiment_test = tests["Sentiment140"]
imdb_test = tests["Movie Database (IMDb)"]
tweets_test = tests["Twitter US Airline Sentiment"]
```

```python
amazon_test.to_json("AMAZON_TEST_LIMPIO.json")
yelp_test.to_json("YELP_TEST_LIMPIO.json")
financial_test.to_json("FINANCIAL_TEST_LIMPIO.json")
stanford_test.to_json("STANFORD_TEST_LIMPIO.json")
sentiment_test.to_json("SENTIMENT140_TEST_LIMPIO.json")
imdb_test.to_json("IMDB_TEST_LIMPIO.json")
tweets_test.to_json("TWEETS_TEST_LIMPIO.json")
```
